```python
# data columns

# videoId,
# title,
# description,
# tags,
# viewCount,
# likeCount,
# favoriteCount,
# commentCount,
# publishedAt,
# channelId,
# channelTitle,
# country

use_columns = ['videoId', 'title', 'description', 'tags', 'viewCount', 'likeCount', 'favoriteCount', 'commentCount', 'publishedAt', 'channelId', 'channelTitle', 'country']
```

```python
from dotenv import load_dotenv
import os
import sys
import pandas as pd
import numpy as np
from youtubeapi import YoutubeApi
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
import matplotlib.font_manager as fm
from utils import utils

# font setting
font_location = 'C:/Windows/Fonts/Malgun.ttf' if sys.platform == 'win32' else '/System/Library/Fonts/AppleSDGothicNeo.ttc'
font_name = fm.FontProperties(fname=font_location).get_name()
# print(font_name)
matplotlib.rc('font', family=font_name)

load_dotenv()
apiKey = os.environ.get('YoutubeApiToken')
url = 'https://youtube.googleapis.com/youtube/v3/'

api = YoutubeApi(apiKey, use_columns)
```

```python
asia_data = pd.read_csv('../data/asia_data_kor.csv')
asia_data['Year'].fillna(method='ffill', inplace=True)
# asia_data
```

# 베트남

- 검색어: `코이티비 베트남`
- 기간: `2018-01-01T00:00:00Z` ~ `2019-01-01T00:00:00Z`

```python
searchQuery = '코이티비 베트남'
country = '베트남'

publishedAfter='2018-01-01T00:00:00Z'
publishedBefore='2019-01-01T00:00:00Z'

datas = api.search_videos(searchQuery, publishedAfter, publishedBefore)
df = api.get_video_dataframe(datas, country)
df
```

```python
select_video_idx = 0

year = df.loc[select_video_idx, 'publishedAtYear']
month = df.loc[select_video_idx, 'publishedAtMonth']

display(year, month)
```

```python
# 2023년 5월까지 기간 데이터 불러오기
period = utils.get_dataframe_by_period(asia_data, year, 1, 2023, 5, country)
```

```python
# 코로나 격리 해제 데이터

exit_covid = pd.read_csv('../data/exit_covid.csv')
exit_covid
```

```python
# 동영상 게시 일자와 코로나 격리해제 일자
pub_date = str(year) + '년'+str(month) + '월'
exit_date = str(exit_covid[exit_covid['국가명'] == country]['year'].values[0]) + '년' + str(exit_covid[exit_covid['국가명'] == country]['month'].values[0]) + '월'
```

```python
# barplot 표시를 위한 color 세팅
edge_colors = []
fill_colors = []

for c in period['date']:
  if c == pub_date:
    edge_colors.append('b')
    fill_colors.append('orange')
  elif c == exit_date:
    edge_colors.append('r')
    fill_colors.append('orange')
  else:
    fill_colors.append('grey')
    edge_colors.append('w')
```

```python
# 증감율 생성
period[f'{country}_diff'] = period[f'{country}_명수'].rolling(window=2).apply(lambda x: ((x.iloc[1] - x.iloc[0])/x.iloc[0]))
period[f'{country}_pct'] = period[f'{country}_명수'].pct_change() * 100

# bar + line
fig, ax1 = plt.subplots(figsize=(25, 10))
plt.title(f'{searchQuery} 전/후 방문객 수 비교', pad=10, fontweight='bold')

plt.xlabel('월별')
sns.lineplot(data = period, x='date', y=f'{country}_pct', ax=ax1)
plt.ylabel('증감')
plt.xticks(rotation=45)
ax2 = ax1.twinx()
sns.barplot(data = period, x='date', y=f'{country}_명수', alpha=0.5, ax=ax2, edgecolor=edge_colors, palette=fill_colors)
plt.ylabel('방문객 수')
plt.tight_layout()

plt.savefig(f'./result_graph/{searchQuery} 전후 방문객 수 비교.png')
plt.show()
```
